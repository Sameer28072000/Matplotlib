Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans:-Scatter plot is used to define relationship between two variable.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
print(x)
```

```python
print(y)
```

```python
plt.scatter(x,y)
plt.xlabel('This is X-axis')
plt.ylabel('This is Y-axis')
plt.title('X VS Y')
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('A single plot')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Ans:-Box plot is help to find outliers in the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()

bp = ax.boxplot(data)

ax.x-label(['Box 1', 'Box 2'])
ax.ylabel('Values')
ax.title('Box Plot')

plt.show()
```
